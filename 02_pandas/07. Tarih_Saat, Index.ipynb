```python
import pandas as pd
```

```python
data_path = "D:\\OneDrive\\egitim\\verilen\\bogazici-compec-python-ml\\Datasets\\"
data_set = "OnlineRetailDirty.xlsx"
```

# Dataframe okurken indeks sütunu belirlemek

```python
# Read dataset. While reading make date column dataframe index and parse date column as timestamp
df = pd.read_excel(data_path+data_set, parse_dates=['InvoiceDate'], index_col="InvoiceDate")
```

```python
df.head()
```

# Dataframe'i Indekse Göre Sıralama

```python
df = df.sort_index()
```

```python
df.head()
```

# Tarih (Index) Bilgisinden Yeni Nitelikler Üretmek

```python
df['year'] = df.index.year
df['day_of_month'] = df.index.day
df['month'] = df.index.month
df['day_of_week'] = df.index.dayofweek
df['quarter'] = df.index.quarter
df['week_of_year'] = df.index.weekofyear
```

```python
df.head()
```

# Tarihlerde bir tuhaflık var mı?

```python
print("max month:", df.month.max())
print("min month:", df.month.min())
print("max year:", df.year.max())
print("min year:", df.year.min())
print("max day of month:", df.day_of_month.max())
print("min day of month:", df.day_of_month.min())
print("max day of week:", df.day_of_week.max())
print("min day of week:", df.day_of_week.min())
```

```python
# Yukarıdaki bilgiye göre bir anormallik yok. Yalnızca kütüphaneyi yazanlar Pazartesi gününü
# 0, Pazar gününü 6 kabul etmişler. O yüzden haftanın günleri 0-6 aralğında.
```

# Tarihlere göre satışları inceleme

```python
# İncelemeye geçmeden önce her bir malzemenin o işlemde ne kadar tuttuğunu hesaplamamız lazım.,
# Çünkü elimizde sadece Quantity ve UnitPrice var. Yani satış miktarı ve birim fiyat.
# Bu ikisinin çarpımıntan TotalPrice hesaplayalım
```

# ToplamFiyatı hesaplama ve yeni sütun olarak ekleme

```python
df['TotalPrice'] = df['Quantity'].mul(df['UnitPrice'])
```

```python
df.head()
```

# Haftanın hangi günlerinde ne kadar satış olmuş?

```python
df[['day_of_week','Quantity']] \
.groupby('day_of_week') \
.agg('sum') \
.sort_values('Quantity', ascending=False) \
.head(10)
```

## Günlerin İsimlerini Yazalım

```python
days = []
for i in df.day_of_week:
    if i == 0:days.append('Monday')
    elif i == 1:days.append('Tuesday')
    elif i == 2:days.append('Wednesday')
    elif i == 3:days.append('Thursday')
    elif i == 4:days.append('Friday')
    elif i == 5:days.append('Saturday')
    else :days.append('Sunday')
```

```python
df['day_name'] = days
```

```python
df.head()
```

# Gün İsimlerine Göre Toplam Satış

```python
df[['day_name','Quantity']] \
.groupby('day_name') \
.agg('sum') \
.sort_values('Quantity', ascending=False) \
.head(10)
```

# Aylara göre ortalama ve toplam satış miktarı ile işlem sayısı

```python
df[['year','month','TotalPrice']] \
.groupby(['month']).TotalPrice \
.agg(['count','mean','sum']) \
.sort_values('sum',ascending=False) \
.head(30)
```

# Bir işlemde gerçekleşen en büyük satış

```python
df[['InvoiceNo','TotalPrice']] \
.groupby(['InvoiceNo']).TotalPrice \
.agg(['max']) \
.sort_values('max', ascending=False) \
.head()
```

```python
# Hangi müşteri bir defada en çok satış yapmış
```

```python
df[['InvoiceNo','TotalPrice','CustomerID']] \
.groupby(['InvoiceNo','CustomerID']).TotalPrice \
.agg(['max']) \
.sort_values('max', ascending=False) \
.head(30)
```
